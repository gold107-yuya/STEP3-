```python
stations = yamanoteline['最寄駅'].unique().tolist()
stations
```

```python
stations[0]
```

```python
yamanoteline[yamanoteline['最寄駅']==stations[0]]["家賃（万円）"].mean()
```

```python
yamanoteline[yamanoteline['最寄駅']==stations[0]]["占有面積"].mean()
```

```python
yamanoteline['間取り'].mode()
```

```python
data = []

for station in stations:
    rentprice_ave = yamanoteline[yamanoteline['最寄駅']==station]["家賃（万円）"].mean()
    area_ave = yamanoteline[yamanoteline['最寄駅']==station]["占有面積"].mean()
    roomtype = yamanoteline[yamanoteline['最寄駅']==station]["間取り"].mode().tolist()

    datum={}
    datum['駅名']=station
    datum['平均家賃']=rentprice_ave
    datum['平均面積']=area_ave
    datum['最多間取り']=roomtype
    
    data.append(datum)
```

```python
data
```

```python
df_ave = pd.DataFrame(data)
df_ave
```

```python
df_ave.sort_values(by='平均家賃',ascending=False)
```

```python
df_ave.to_csv('yamastation_ave.csv')
```
